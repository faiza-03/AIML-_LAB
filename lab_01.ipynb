##ACTIVITY-02:DATA VISUALIZATION

EXE-01:SCATTERPLOT-BASIC

```python
people=['arun','david','akshaya','vinisha','vinuthna','sathwika']
age=[15,18,20,18,22,12]
weight=[50,55,70,48,70,40]
height=[170,183,188,166,177,167]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.title("peoples weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='pink',edgecolor='k',alpha=0.75,bins=5)
plt.xlabel("histogram of patient weight",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.xticks(fontsize=15,rotation=30)
plt.yticks(fontsize=15)
plt.show()
```
